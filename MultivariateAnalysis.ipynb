IMPORTING THE LIBRARIES

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

IMPORTING THE DATASET

```python
data=pd.read_csv("Data.csv")
```

```python
data.head()
```

```python
data.columns=["V"+str(i) for i in range(1,len(data.columns)+1)]
```

```python
data.head()
```

```python
data.info()
```

SUMMARY STATISTICS FOR MULTIVARIATE DATA

```python
X=data.loc[:,"V2":]
X.apply(np.mean)
```

```python
X.apply(np.std)
```

```python
X.apply(np.max)
```

```python
X.apply(np.min)
```

```python
y=data.V1
d1= X.groupby(y)
display(d1.apply(np.mean))
```

```python
display(d1.apply(np.std))
```

PLOTTING MULTIVARIATE DATA

MATRIX SCATTERPLOT

```python
pd.plotting.scatter_matrix(data.loc[:,"V2":"V14"], diagonal="kde", figsize=(15,15))
plt.show()
```

PROFILE PLOT

```python
ax=data[["V2","V3","V4","V5","V6"]].plot(figsize=(15,15))
```
