IMPORTING THE LIBRARIES

```python
import pandas as pd
import numpy as np
```

IMPORTING THE DATASET

```python
df = pd.read_csv('iris.csv')
df
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

VISUALIZING UNIVARIATE CONTINUOUS DATA

1. UNIVARIATE SCATTER PLOT

```python
plt.scatter(df.index, df['sepal.length'])
plt.show()
```

```python
sns.scatterplot(x = df.index, y= df['sepal.length'], hue= df['variety'])
```

2.LINE PLOT

```python
plt.figure(figsize = (5,5))
plt.title('Line plot of petal length')
plt.xlabel('index', fontsize = 15)
plt.ylabel('petal length', fontsize = 15)
plt.plot(df.index, df['petal.length'], markevery = 1, marker = 'd')
for name, group in df.groupby('variety'):
  plt.plot(group.index, group['petal.length'], label = name, markevery = 1, marker = 'd')
plt.legend()
plt.show()
```

```python
sns.set(rc={'figure.figsize':(5,5)})
sns.set(font_scale = 1.5)
fig= sns.lineplot(x = df.index, y = df['petal.length'], markevery = 1, marker ='d', data = df, hue= df['variety'])
fig.set(xlabel = 'index')
```

3.HISTOGRAMS

```python
plt.hist(df['petal.width'])
```

```python
sns.distplot(df['petal.width'], kde = False, color = 'black', bins = 10)
```

4. DENSITY PLOTS

```python
df['petal.length'].plot(kind = 'density')
```

```python
sns.kdeplot(df['petal.length'], shade = True)
```

5. BOX PLOTS

```python
plt.boxplot(df['sepal.width'])
```

```python
df2 = df.drop('variety', axis = 1)
plt.title('Box plots of variables')
plt.boxplot(df2.values, labels = ['SL','SW','PL','PW'])
```

```python
sns.boxplot(df['sepal.width'])
```

```python
sns.boxplot(x = "variable", y ="value", data = pd.melt(df2))
```
